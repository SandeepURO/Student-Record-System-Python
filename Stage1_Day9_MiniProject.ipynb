A system that can:

➕ Add a student
📋 Show all students
🔍 Search a student by name
❌ Delete a student
📊 Show class average marks

```python
students=[] # first we need a empty list for inserting(add) or deleting(remove)
```

```python
# creating a function to adding students in the empty list
def add_student(name,age,branch,marks):
    student={
        "name":name,
        "age":age,
        "branch":branch,
        "marks":marks
    }
    students.append(student)
    print(f"{name} added successfully!!!")
```

```python
add_student("sandeep",20,"CSE",99)
add_student("Rahul",22,"ECE",99)
add_student("Harsha",18,"CSE",99)

print(students)
```

```python
# made a function to display all the students in the list
def show_all():
    if len(students)==0:
        print("No students found!!")
    else:
        print("\n--- All Students ---")
        for i, student in enumerate(students, 1):
            print(f"{i}. {student['name']} | age: {student['age']} | branch: {student['branch']} | Marks: {student['marks']}")
```

```python
show_all()
```

```python
# made a function for searching a student in list
def search_student(name):
    for student in students:
        if student['name'].lower() == name.lower():
            print(f"\n---- Student Found ----")
            print(f"Name:{student['name']}")
            print(f"Age:{student['age']}")
            print(f"Branch:{student['branch']}")
            print(f"Marks:{student['marks']}")
            return
    print(f"Student '{name}' not found in the Recorded")
        
```

```python
search_student("sandeep")
search_student("Amit")
```

```python
# made a function for deleting a student in the list 
def delete_student(name):
    for student in students:
        if student["name"].lower()==name.lower():
            students.remove(student)
            print(f"{name} deleted successfully!!!")
            return 
    print(f"student '{name}' not found!!")
    
```

```python
delete_student("Rahul")   # exists
show_all()                # verify Rahul is gone
delete_student("Amit")    # doesn't exis
```

```python
# made a function for the class analysis
def cls_avg(students):
    if len(students) == 0:
        print("No students Founded")
        return
    total=0
    for student in students:
        total = total+student["marks"]
    average = total/len(students)
    highest= max(students, key=lambda s: s["marks"])
    lowest= min(students, key=lambda s: s["marks"])
    print(f"\n--- Class Statistics ---")
    print(f"Average Marks: {average}")
    print(f"Highest : {highest['name']} with {highest['marks']}")
    print(f"Lowest : {lowest['name']} with {lowest['marks']}")
    
```

```python
cls_avg(students)
```

```python
add_student("Amit", 21, "ME", 65)
cls_avg(students)
```

```python
# This function is for the MAIN MENU for the whole List
def main_menu():
    while True:
        print("\n ===== Student Record System =====")
        print("1. Add Student")
        print("2. Show All Student")
        print("3. Search Student")
        print("4. Delete Student")
        print("5. Class Statistics")
        print("6. EXIT")
        choice=input("Enter choice (1-6): ")
        if choice == "1":
            name = input("Name:").strip()# here we are using "strip()" because if any input-user uses extra space accidentally, strip() can removes extra space 
            age = int(input("Age:"))
            branch = input("Branch:").strip()
            marks = float(input("Marks:"))
            add_student(name,age,branch,marks)
        elif choice=="2":
            show_all()
        elif choice=="3":
            name=input("\nEnter name to search: ").strip()
            search_student(name)
        elif choice=="4":
            name=input("\nEnter the name to delete: ").strip()
            delete_student(name)
        elif choice=="5":
            cls_avg(students)
        elif choice=="6":
            print("\nThanks For Visiting!!!")
            break
        else:
            print("\nInvalid choice!!!")
        
        
```

```python
main_menu()
```
